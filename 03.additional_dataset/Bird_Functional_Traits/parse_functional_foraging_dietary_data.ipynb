```python
raw_data = pd.read_csv('./01.functional_traits/AVONET_Raw_Data.csv')
raw_data.columns
```

```python
raw_data = raw_data.groupby(['Species1_BirdLife', 'Species2_eBird',
       'eBird.species.group', 'Species3_BirdTree']).first().reset_index(drop=False)[['Species1_BirdLife', 'Species2_eBird',
       'eBird.species.group', 'Species3_BirdTree']]
```

```python
raw_data
```

```python
file_path = "./01.functional_traits/41559_2019_1070_MOESM3_ESM.csv"
functional_data = pd.read_csv(file_path)
functional_data['Binomial'] = [i.replace("_",' ') for i in functional_data['Binomial'].values]
```

```python
functional_data
```

```python
functional_data = functional_data[['Binomial','PC1','PC2','PC3','PC4','PC5','PC6','PC7','PC8','PC9',
                                   'Beak_PC1','Beak_PC2','Beak_PC3','Beak_PC4',
                                   
             'TrophicLevel','TrophicNiche','ForagingNiche']].rename(columns={
    'Binomial':'species'
})
functional_data.head()
```

```python
ebird_name = pd.read_csv('./03.ebird_name/eBird_Taxonomy_v2021.csv')
ebird_name = ebird_name[ebird_name['CATEGORY']=='species']
ebird_name = ebird_name[['PRIMARY_COM_NAME','SCI_NAME','ORDER1','FAMILY']].rename(columns={
    'SCI_NAME':'species',
    'PRIMARY_COM_NAME':'common_name',
    'ORDER1':'order',
    'FAMILY':'family'
})
ebird_name.head()
```

```python
birdlife_data = pd.read_csv('./01.functional_traits/AVONET1_BirdLife.csv')
birdlife_data = birdlife_data[['Species1','Beak.Length_Culmen', 'Beak.Length_Nares', 'Beak.Width', 'Beak.Depth',
       'Tarsus.Length', 'Wing.Length', 'Kipps.Distance', 'Secondary1',
       'Hand-Wing.Index', 'Tail.Length', 'Mass','Habitat', 'Habitat.Density', 'Migration',
                              'Primary.Lifestyle','Range.Size']].rename(columns={
    'Species1':'Species1_BirdLife'
})


birdlife_data.head()
```

```python
inno = pd.read_csv('./01.functional_traits/Innovativeness.csv')
inno = inno[['Species','TotalInnovations', 'TotalFoodTypeInnovations',
       'TotalTechnicalInnovations', 'InnovationYesOrNo',
       'FoodTypeInnovationYesOrNo', 'TechnicalInnovationYesOrNo']]
inno['Species'] = [" ".join(i.split("_")) for i in inno.Species.values]
inno = inno.rename(columns={'Species':'Species1_BirdLife'})
inno.columns
```

```python
brain_size = pd.read_csv('./01.functional_traits/Brain_size.csv')\
    [['Species','BrainSize','RelativeBrainSize']].rename(columns={'Species':'Species1_BirdLife'})

brain_size['Species1_BirdLife'] = [" ".join(i.split("_")) for i in brain_size['Species1_BirdLife'].values]
```

```python
brain_size
```

```python
#### merge
merged = pd.merge(
    pd.merge(
        pd.merge(ebird_name, raw_data, left_on='species',right_on='Species2_eBird',how='left'), 
        birdlife_data, left_on='Species1_BirdLife', right_on = 'Species1_BirdLife', how='left'
    ),
    functional_data, on='species',how='left'
)

merged = pd.merge(
    pd.merge(merged,
                 inno,on='Species1_BirdLife', how='left'),
    brain_size, on='Species1_BirdLife', how='left'
)


merged.head()
```

```python
merged.isnull().sum()
```

```python
merged.to_csv('functional_trophical_foraging_data.csv',index=False)
```
